##UML Summary

* <font color=#0000ff> **Class–Diagrams**</font> show the static properties of classes and the (possible) connections between them.
* <font color=#0000ff> **Object–Diagrams**</font> show the concrete properties and connections for objects as a <font color=#ff0000> **snapshot**</font> for a specific scenario.
* <font color=#0000ff> **Sequence Diagrams**</font> show the chronological sequence, the communication of the objects involved for a specific scenario.

That's not all: There are also <font color=#008000> *use-case diagrams, collaboration diagrams, package diagrams, ...*</font>

see e.g.: [UML Quick Reference](http://www.holub.com/goodies/uml/index.html)

---

For large OO projects with professional teams, <font color=#0000ff> **formal UML design**</font> plays an important role: 30-60% of the effort.
* Extensive training required
* Powerful commercial tools for UML design: <font color=#008000> *Rational Rose, Together, ...*</font>
* Design in all details in UML diagrams $\Rightarrow$ automatic generation of source code from it
* analogous to <font color=#0000ff> **re-engineering**</font> : Tools analyze existing code and create UML diagrams
  

But also very useful on a smaller <font color=#008000> *(projects in physics)*</font> scale:
* UML diagrams as simple sketches by hand or with <font color=#0000ff> **Umbrello**</font> program <font color=#008000> *(Linux-tool)*</font>
* Good basis for discussion <font color=#008000> *(cf source code)*</font> and for documentation
* The best way to precisely capture the requirements is to discuss <font color=#0000ff> **use–cases**</font> $\Rightarrow$ <font color=#0000ff> **Sequence-Diagrams**</font>
