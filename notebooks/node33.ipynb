Typical example of code aging and sort of <font color=#008000> *rotten*</font>.
* simple, clear function at the beginning
* more and more features will be added
* backward–compatibility forces obscure constructions

OO Analysis and Design:
* Flexible, easily expandable systems
* Well-defined <font color=#008000> *responsibility*</font> of a class/function.
