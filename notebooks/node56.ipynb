## XML structure in detail

* XML document begins with XML markup:
```xml
<?xml version="1.0" encoding="ISO-8859-1"? >
```
Specifies the XML version and the character encoding, if any
* The text range from `<Class>` to `</Class>` is a <font color=#008000> *XML element*</font>. Elements must have opening and closing tags (possibly combined in one) (`<Class>` or `</Class>` or `<img />`)
* Elements can contain other elements, like the <font color=#0000e6> ``Student``</font> element in the example <font color=#0000e6> ``Grade``</font> element.
* An element can also have attributes, such as <font color=#0000e6> ``Student``</font> and the attribute <font color=#0000e6> ``(name="Paul")``</ font>
* Order of sub-items doesn't matter...
* Flexibly expandable, additional elements can be inserted, e.g. <font color=#0000e6> ``BachelorGrade``</font> , existing applications continue to run

<font color=#008000> *(in contrast to other formats like ASCII Text, CSV, Pickle, SQL, ...)*</font>

---

**Character Data and Markup**
* Markup tags: starts with <font color=#0000e6> ``<``</font> or <font color=#0000e6> ``&``</font> and ends with <font color=#0000e6> ` `>``</font> or <font color=#0000e6> ``;``</font>
* Character Data: Any character string in between. Must not contain the start of a markup tag directly.
* XML Text - Markup and Character Data together form the XML Text and together with the Declaration Header the XML Document. The markup is a kind of schema for the interpretation of the character data
* Comments can be inserted as `<!-- comment-text -->`

---

**Item Tag**
* The element tag contains or defines its type and structure
* A start tag can contain attributes, the order does not matter: `<Value a="12" b="16"/>` is the same as `<Value b="16" a="12"/> `.
* Attribute values ​​must be in quotes: <font color=#0000e6> ``attribute="value"``</font>
* The end tag must match the name of the start tag exactly

---

**Document Types and Schemas**
* Document type is a type of data definition for an XML document
* Specifies the allowed structure - similar to how a class and its member variables specify the data structure
* ... is also called <font color=#008000> *Scheme*</font>
* One possibility to define this is the <font color=#0000ff> **DTD**</font> : Document Type Definition - but there are still a number of alternatives

---

**DTDs**
* Can be integrated into the document as a header or separately as a <font color=#0000e6> ``.dtd``</font> file
* DTDs provide a kind of type definition for documents that specifies the correct structure
* This can be used to validate an XML document, some XML parsers do this automatically.
* we limit ourselves to simple examples ...

---

Built-in DTD:

```xml
<?xml version="1.0"?> 
<!DOCTYPE product [ 
<!ELEMENT name (#PCDATA)> 
<!ELEMENT price (#PCDATA)> 
<!ELEMENT product (name, price)> 
]> 
<product> 
<name>Bean Crusher</name> 
<price>3.95</price> 
</product>
```
