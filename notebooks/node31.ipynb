## Software engineering
Central branch of computer science deals with <font color=#008000> *Software Engineering*</font> . Original approach:
* Methods and procedures from other technical areas transferred to software development: Proven engineering processes from large-scale plant construction, construction industry, vehicle technology, ...
* Accurate, detailed specification
* Meticulous design
* Division of the project into independent standard components as much as possible
* Many important advances in software engineering in recent decades
* But the goal of being able to adequately control software projects analogously to other technical areas with <font color=#008000> *Engineering*</font> methods was not achieved:
<font color=#008000> *Only about 20% of the software projects are successful in the narrower sense, i.e. budget, deadlines, functionality largely or fully fulfilled*</font> (Source: iX Magazin 2/06)
* Software to be dynamic and lively; Software in use moves in a highly complex state space; Effects cannot be recorded in detail

---

The way of programming extremely depends on the size of the project <font color=#008000> *(note: rule of thumb is 10 lines of program code per day!)*</font>
* **O(50-500) lines**
* Courses, mini-projects. Usually solvable without a formal design process:
* think for a moment, then program. Program fully manageable by individual. Commented, structured code can also be used by others <font color=#008000> *(...but mostly it's completely new)*</font>.
* **O(500-10k) rows**
* To be managed by individuals or small groups. Further development and dissemination critical without formal design.
* **O(100k-1M) rows**
* Development team required. Division into sub-projects. Detailed design crucial, possibly bulk of the time.
* **> 1M rows**
* often fail... <font color=#0000ff>
    

**Design effort increases non-linearly with the size of the project!**</font>

---

<font color=#0000ff> **Problems with large software projects:**</font>

**Software is <font color=#ff0000> **rigid**</font> or <font color=#ff0000> **fragile**</font>** , i.e.
* Difficult to see dependencies
* Small changes have big side effects
* Unpredictable side effects occur with every change in the process or exchange of modules
* Hard to estimate effort for modifications
* If $P(bug\vert change) \approx 1$ the system can no longer be further developed.
* Management policy <font color=#008000> *“Don't touch a working system”*</font>

---

**Software is <font color=#ff0000> **not reusable**</font>** , i.e.
* Code already exists to solve a problem
* But contains dependencies on many other things irrelevant to the problem
* some minor adjustments needed
* 2 options:
* You take responsibility for the code and the necessary adjustments
* Make independent <font color=#008000> *cut& paste*</font> copy of needed code.
* Ideally, you just want to <font color=#008000>use</font> the code and not <font color=#008000>take responsibility for maintenance</font>.

---

**<font color=#ff0000> **Dependency Management**</font>**
* A large number of dependencies between the individual components are unavoidable in a large software project
* Controlling the dependencies of the individual components is crucial
* OO methods to monitor dependencies, e.g. UML diagrams
* OO “tricks” to minimize dependencies <font color=#008000> *(abstract classes, interfaces)*</font>
