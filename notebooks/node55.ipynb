## XML Basics
 

XML (eXtensible Markup Language) is the same concept: the content is classified ("marked up") using <font color=#0000ff> **Tags**</font>.

However, the tags in XML are not specified, but can be freely defined (=eXtensible).
* Universal standard for data exchange between different applications (application neutrality):
<font color=#008000> *Create with Java and process with Python, ...*</font>
Regardless of platform (Unix, Windows, Mac, ...)
* Extensible without having to adapt existing applications
* Directly readable and editable with simple text editor (more or less...)

First simple example:
```xml
<?xml version="1.0" encoding="ISO-8859-1"? > 
<Class> 
<Student name="Mueller" prename="Max"> 
<ID>123498765</ID> 
<Grade>2.0</Grade> 
</Student> 
<Student name="Kunz" prename="Maria"> 
<ID>123498762</ID> 
<Grade>3.0</Grade> 
</Student> 
</Class> 
<Student name="Hinz" prename="Karl"> 
<ID>12349879</ID> 
<Grade>3.3</Grade> 
</Student> 
</Class>
```

Very similar to HTML, but you can define data fields or tags yourself.
