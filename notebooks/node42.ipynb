## Simple timer
Timers are simple and flexible methods of measuring program runtime. The following example demonstrates a simple timer using a decorator:

```python
import time

def timefunc(f):
    def f_timer(*args, **kwargs):
        start = time.time()
        erg = f(*args, **kwargs)
        ende = time.time()
        print(f.__name__, 'brauchte', ende - start, 's')
        return erg
    return f_timer

def get_number():
    for x in range(50000000):
        yield x

@timefunc
def teure_funktion():
    for x in get_number():
        i = x ^ x ^ x
    return 'Ergebnis'

res = teure_funktion()
```

Timers require incrementally testing higher-level functions and then digging deeper into individual functions to identify the bottleneck before it can be fixed.

The [timeit](https://docs.python.org/2/library/timeit.html) module is also very useful for testing small program snippets.
