## Data I/O - Summary

Many possibilities and technologies for data I/O.
 

No general solution, select according to problem and requirements.

* ASCII text output or CSV format:
<font color=#0000ff> **Pro:**</font> Simple, portable.
<font color=#ff0000> **Con:**</font> No structure, no complex data.
* Binary format like Pickle:
<font color=#0000ff> **Pro:**</font> Fast, efficient, storage of complex object structures.
<font color=#ff0000> **Con:**</font> Python-specific
* (Relational) database (SQLite, Postgres, Oracle, ...):
<font color=#0000ff> **Pro:**</font> Storage of any complex data, links between tables.
<font color=#ff0000> **Con:**</font> Bulky syntax, no direct IO of objects.
* Structured output with XML or JSON:
<font color=#0000ff> **Pro:**</font> Complex structures possible, readable (and editable) format, widespread standard for data exchange
<font color=#ff0000> **Con:**</font> Inefficient, memory intensive, no direct IO of objects
