# Profiling of programs
> “We should forget about small efficiencies, say about 97% of the time: premature optimization is the root of all evil. Yet we should not pass up our opportunities in that critical 3%.” ([Donald Knuth](http://en.wikiquote.org/wiki/Donald_Knuth)) <!-- quoted from https://softwareengineering.stackexchange.com/questions/80084/is-premature-optimization-really-the-root-of-all-evil -->

One should not normally program a program for the highest efficiency and speed in the first attempt, as this usually quickly makes the code confusing and difficult to debug. That means you should first make sure that you have a working program that meets the specifications <font color=#008000> *( = tests!)*</font> and only then examine the performance in tests under real conditions and optimize if necessary. There are special <font color=#0000ff> **profiling**</font> methods that make systematic optimization or the search for <font color=#008000> *bottlenecks*</font> easier.
