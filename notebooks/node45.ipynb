## Profiling in IPython
[IPython](https://ipython.org/) (Interactive Python) is a package that builds on top of Python and, among many other things, extends the command line environment.

The profiler or the `line_profiler` can also be used within IPython. The <font color=#0000e6> ``%prun``</font> command can be used to run the profiler on a previously loaded function:

```python
import numpy as np
import numpy.random as rdn

def test():
    a = rdn.randn(100000)
    b = rdn.randn(100000)
    c = a * b

%prun test()
```

The `line_profiler` can be used like this:

```python
%load_ext line_profiler
%lprun -f test test()
```
