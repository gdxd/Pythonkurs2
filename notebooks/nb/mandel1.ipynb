```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot of Mandelbrot Set
# naive version with standard python loop and lists
# iteration limit for mandelbrot convergence
max_iter = 100

# dimension of plot
h = 400
w = 400

# x range and step
xlow = -1.5
xhi = 0.5
xstep = (xhi-xlow)/w

# y range and step
ylow = -1.
yhi = 1.
ystep = (yhi-ylow)/h

img = [] # empyt list
y = ylow
for i in range(h):
    x=xlow
    imgx = w*[0] # pixel column
    for j in range(w):
        c = complex( x, y)
        z = complex( 0, 0)
        for k in range(max_iter):
            # mandelbrot algorithm
            z = z*z + c 
            if (abs(z) > 2.):
                break
        imgx[j] = k # use #-iterations as pixel value
        x += xstep
    img.append(imgx)
    y += ystep
    
plt.imshow(img)
#plt.imshow(img, cmap=plt.cm.jet)
pass
```

```python
max_iter = 100
h, w = 400, 400

img = np.zeros((h, w)).astype('int')
for i, real in enumerate(np.linspace(-1.5, 0.5, w)):
    for j, imag in enumerate(np.linspace(-1, 1, h)):
        c = complex(real, imag)
        z = 0 + 0j
        for k in range(max_iter):
            z = z*z + c
            if abs(z) > 2:
                break
        img[j, i] = k

plt.grid(False)
plt.imshow(img, cmap=plt.cm.jet)
pass
```

```python
def mandel(c, z=0, max_iter=100):
    for k in range(max_iter):
        z = z*z + c
        if abs(z) > 2:
            return k
    return k
```

```python
def mandelbrot(w, h, xl=-1.5, xu=0.5, yl=-1, yu=1):
    img = np.zeros((h, w)).astype('int')
    for i, real in enumerate(np.linspace(xl, xu, w)):
        for j, imag in enumerate(np.linspace(yl, yu, h)):
            c = complex(real, imag)
            img[j, i] = mandel(c)
    return img
```

```python
img = mandelbrot(w=400, h=400)
plt.grid(False)
plt.imshow(img, cmap=plt.cm.jet)
pass
```
