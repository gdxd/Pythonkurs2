## Classes in UML
Classes describe objects:
* Status: values ​​of member variables
* Constructor: create the object
* Interface: Signature of the member functions
* Behavior: Implementation of member functions

Classes or objects have <font color=#0000ff> **relationships**</font> to other classes/objects
