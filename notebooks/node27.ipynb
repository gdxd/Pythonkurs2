## Relationships between classes

Classes or objects can depend on each other in many ways:
* <font color=#0000ff> **Association:**</font> Unspecified (usually loose) coupling of classes, only in one direction or bi-directional
* <font color=#0000ff> **Aggregation or Composition:**</font> A class contains one (or more) objects of another class.
* <font color=#0000ff> **Inheritance**</font>

---

**Binary Association**
 

Both classes know each other...

![Image BinAssoc](figures/UML_BinAssoc.png "Image BinAssoc") ... as a reference or function argument or in container or ...

---

**More common is the Unary Association:**

<font color=#0000ff> **A**</font> knows <font color=#0000ff> **B**</font> but <font color=#0000ff> **B**</font> don't know about <font color=#0000ff> **A**</font>

![Image UnAssoc](figures/UML_UnAssoc.png "Image UnAssoc")

---

**Aggregation/Composition:**

<font color=#0000ff> **A**</font> contains <font color=#0000ff> **B:**</font> <font color=#008000> *(whole–part relation)*< /font> , e.g. car contains engine

![Image Composition](figures/UML_Composition.png "Image Composition")Subtle difference between aggregation and composition, the latter controls <font color=#008000> *lifetime*</font> of the contained object $\Rightarrow$ see literature

---

**Association details:**

* Association can have name <font color=#008000> *(uses)*</font>
* Names for the <font color=#0000ff> **Roles**</font> of the two classes can be at the ends <font color=#008000> *(Master, Slave)*</font> ...
* ... and the multiplicity, i.e. the number of objects involved
![Image UnAssoc2](figures/UML_UnAssoc2.png "Image UnAssoc2")
* If read correctly, this results in a compact screenplay:
<font color=#008000> *One .. Master .. uses .. one or more .. Slaves*</font>

**Summary:**

UML class diagrams can be used to graphically represent properties of classes, i.e. <font color=#008000> *Name, member variables, interfaces and methods*</font> in a compact manner.

Depending on the context, one can output a concise listing of all member variables and methods with arguments, or only an incomplete list of the relevant elements.

One can express different kinds of relationships between classes with UML:

<font color=#008000> *Association, aggregation, composition, inheritance*</font>

The links can be vague sketches or precise information about <font color=#008000> *names, roles, multiplicities, ...*</font>.

Class diagrams describe the <font color=#ff0000> **static design**</font> of the program structure.
