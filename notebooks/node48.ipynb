### Introduction

Depending on the problem, data storage can be relatively simple, but also arbitrarily complex. A few illustrative examples:
* Simple series of measurements of numeric values: direct storage as a list of numbers, reading into the program as a corresponding array/list.
* Tabular form: Each entry (=line) contains several elements, any number and type, but fixed. E.g. student data (name, subject, grades, ...). Save as ASCII file or spreadsheet or csv file. Or even XML/JSON if flexible extensibility is needed.
Read in/use in program e.g. as 2D list or list of dicts.
* Nested tables and cross-references: E.g. data for Facebook account: personal data, list of posts, list of comments, list of friends, ...
Depending on the application/complexity: XML/JSON, Python-pickle, SQL Database, ...
